```python
# open patent data 
df = pd.read_csv(R'X:\5_Research\Paul\data\genes_of_technology\full_patent_data_lens_marine_2689_familychecked_fulltxts_keywordsentences_PVE.csv', index_col = 0)
    
```

```python
# format for sampling
keyword_index = [keyword for keywords in df['keyword_list'] for keyword in ast.literal_eval(keywords)]
sentences = [sentence for keywords, sentence in zip(df['keyword_list'], df['sentence']) for _ in ast.literal_eval(keywords)]
ids = [patent_id for keywords, patent_id in zip(df['keyword_list'], df['patent_id']) for _ in ast.literal_eval(keywords)]
```

```python
df_kw = pd.DataFrame(zip(keyword_index, sentences, ids), columns = ['keyword', 'sentence', 'patent_id'])
```

```python
#oversample small categories
max_size = df_kw['keyword'].value_counts().max()
lst = [df_kw]
for class_index, group in df_kw.groupby('keyword'):
    lst.append(group.sample(max_size-len(group), replace=True))
df_kw_oversample = pd.concat(lst).set_index('keyword')
```

```python
#presample for 13 sentences to get inform the GPT prompt
df_kw_presample = df_kw_oversample.sample(n=13, random_state = 1)
list(df_kw_presample['sentence'])
```

```python
# get the sample for the GPT labelling
df_kw_finalsample = df_kw_oversample.sample(n=133, random_state = 2)
list(df_kw_finalsample['sentence'])
```

```python
countlist = Counter(list(df_kw_finalsample.index))
countlist.most_common()
```

```python
#GPT
nr = 0
patent_ids = []
sentences = []
responses = []
keywords = []
for sentence in list(df_kw_finalsample['sentence']):
            my_chat_bot = "You are a helpful assistant that categorizes the sentences into 'Public Value Expression' (PVE) or 'No PVE'. If the sentence is categorized as ‘PVE’, further classify it into Direct PVE or ‘Contextual PVE’. Use the following definitions: Public value expression (PVE): A public value expression is one that indicates societal benefits promised to or for people or ecosystems. Societal benefits for people includes, among other aspects, potential enhancement of safety, security, justice, equality, human rights, and global health. Societal benefits for ecosystems include, among other aspects, enhancements to sustainability, nature and the environment. Enhancement includes both positive improvement and the mitigation or prevention of adverse conditions. Public value is based on the creation of shared utility for users, and people or nature more broadly. No Public Value: Instances that refer to technical aspects without any explicit indication of broader societal benefits are the opposite of PVE, as they indicate benefits only accruing to private beneficiaries. Furthermore, do not code a sentence as 'PVE' if it only refers to improving user experience. User experience enhancement does not imply public value expression. Avoid assumptions about possible positive or negative societal, economic, or environmental spillovers if they are not explicitly stated. Direct PVE: Code a sentence as a 'Direct PVE' if the sentence states that the invention being patented explicitly addresses one or more PVEs (see definition of PVE). Contextual PVE: Code a sentence as a 'Contextual PVE' if the sentence mentions a contextual situation that allows us to infer a concern or intention to create or protect PVEs. The sentence must be explicit in its reference to a public value or concern about a societal, environmental, or ethical concern. However, it is contextual because it is not explicitly connected to the invention being patented. Further instructions: The format of your response should be the following: first, assign one of the 'PVE labels' ('No PV'’,  'Direct PVE', or 'Contextual PVE'). Second, add one or two sentences of easy-to-understand explanation on why you assigned the specific label. The explanation must not be longer than two sentences maximum. Third, after the explanation, add a 'PVE type label' for the type of public value expressed (e.g. 'health', 'nutrition', 'economic prosperity', 'sustainability', ...). If no public value has been expressed just add 'No PVE' again. Always separate 'PVE label', explanation, and the 'PVE type label' at the end of the sentence by ';'."
            query = f"Does the following sentence include a 'Direct PVE', a 'Contextual PVE', or 'No PVE'? The sentence is: {sentence[:4096]}"
            
            response = client.chat.completions.create(
                model="gpt-4-turbo-preview",
                messages = [{"role": "system", "content": my_chat_bot},
                            {"role": "user", "content": query}],
            max_tokens=200,
            )
            keywords.append(list(df_kw_finalsample.index)[nr])
            patent_ids.append(list(df_kw_finalsample['patent_id'])[nr])
            sentences.append(list(df_kw_finalsample['sentence'])[nr])
            responses.append(response.choices[0].message.content)
            nr = nr+1
            
```

```python
df_analysed = pd.DataFrame(zip(keywords, patent_ids, sentences, responses), columns = ['keyword','patent_id', 'sentence', 'GPT response'])
```
